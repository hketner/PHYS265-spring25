**Part 1**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xm = 0
ym = 0

def grav_potential(G,M,x,y):
    potential = np.where(np.sqrt((xm-x)**2+(ym-y)**2) <= 0, np.nan,((-1)*(G*M))/np.sqrt((xm-x)**2+(ym-y)**2))
    return np.abs(potential)

G = 6.67e-11
M = 5.9e24
D = 3.8e8
x = np.linspace(0,1.5*D,100)
y = 0

values = grav_potential(G,M,x,y)

fig, ax1 = plt.subplots()
ax1.plot(x,values)
ax1.set_yscale("log")
ax1.set_xlabel("Distance (m)",fontsize=15)
ax1.set_ylabel("Gravitational Potential (J/kg)",fontsize=15)
ax1.set_title("Gravitational Potential vs. Distance")
```

**Part 2**

```python
xm = 0
ym = 0
G = 6.67e-11
ME = 5.9e24
MM = 7.3e22
D = 3.8e8
x = np.linspace((-1.5*D),1.5*D,100)
y = np.linspace((-1.5*D),1.5*D,100)
xe = D/np.sqrt(2)
ye = D/np.sqrt(2)

def potential_sum(G,M,x,y):
    potential_moon = (-(G*MM)/np.sqrt((xe-x)**2+(ye-y)**2))
    potential_earth = (-(G*ME)/np.sqrt((xm-x)**2+(ym-y)**2))
    potentialsum = np.where(((np.sqrt((xm-x)**2+(ym-y)**2))==0)|(np.sqrt((xe-x)**2+(ye-y)**2)==0),np.nan,potential_earth+potential_moon)
    return potentialsum

xmeshgrid,ymeshgrid = np.meshgrid(x,y)

values = np.abs(potential_sum(G,M,xmeshgrid,ymeshgrid))
fig, ax1 = plt.subplots()
mesh = ax1.pcolormesh(xmeshgrid,ymeshgrid,np.log(values))
ax1.set_xlabel("Distance (m)",fontsize=15)
ax1.set_ylabel("Gravitational Potential (J/kg)",fontsize=15)
ax1.set_title("Gravitational Potential vs. Distance")
ax1.set_aspect("equal")
cbar = fig.colorbar(mesh, norm='log')
cbar.set_label("Gravitational Potential Density")
```

```python
fig_contour, ax_contour = plt.subplots(1,1)

lvl = np.linspace(-50,80,250)
mesh = ax_contour.contour(xmeshgrid,ymeshgrid,np.log(values),levels=lvl)
cbar = fig_contour.colorbar(mesh, norm='log')
ax_contour.set_aspect("equal")
ax_contour.set_xlabel("Distance (m)",fontsize=15)
ax_contour.set_ylabel("Gravitational Potential (J/kg)",fontsize=15)
ax_contour.set_title("Gravitational Potential vs. Distance")
ax_contour.set_aspect("equal")
cbar.set_label("Gravitational Potential Density")
```

**Part 3**

```python
G = 6.67e-11
xm = 0
ym = 0
G = 6.67e-11
ME = 5.9e24
MM = 7.3e22
D = 3.8e8
x = np.linspace((-1.5*D),1.5*D,100)
y = np.linspace((-1.5*D),1.5*D,100)
xe = D/np.sqrt(2)
ye = D/np.sqrt(2)

def grav_force(M1,M2,X1,Y1,X2,Y2):
    r = np.sqrt((X2-X1)**2+(Y2-Y1)**2)
    Fx = np.where(X2-X1==0,np.nan,(-G*(M1*M2)*(X2-X1)/r**3))
    Fy = np.where(Y2-Y1==0,np.nan,(-G*(M1*M2)*(Y2-Y1)/r**3))
    return Fx, Fy

xmeshgrid,ymeshgrid = np.meshgrid(x,y)

xvalues,yvalues = grav_force(ME,MM,xm,ym,xmeshgrid,ymeshgrid)

fig, ax_stream = plt.subplots(1,1)
colors = np.log(np.hypot(xvalues,yvalues))
mesh = ax_stream.streamplot(xmeshgrid,ymeshgrid,xvalues,yvalues, density = 0.5, color=colors)
cbar = fig.colorbar(mesh.lines)
ax_stream.set_aspect("equal")
ax_stream.set_xlabel("X (m)",fontsize=15)
ax_stream.set_ylabel("Y (m)",fontsize=15)
ax_stream.set_title("Gravitational Force vs. XY Position")
ax_stream.set_aspect("equal")
cbar.set_label("Gravitational Force Density")
```

```python
import scipy
from scipy.integrate import quad

m0 = 2.8e6
mf = 7.5e5
n = 1.3e4
g = 9.81
ve = 2.4e3

T = (m0 - mf)/n

def Tsi_func(t,m0,n,mf,g,ve):
    m_t = m0 - n*t
    
    if m_t <= 0:
        return 0
        
    vel_change = ve*np.log(m0/m_t) - g*t
    return vel_change 

h,error = quad(Tsi_func,0,T,args=(m0,n,mf,g,ve))

print("Burn Time: ", T, "s")
print("Altitude: ",h, "m")
```
